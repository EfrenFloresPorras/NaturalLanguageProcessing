```python
# reading text
with open('lincoln_speech.txt', 'r') as file:
    text = file.read()
```

```python
# stopwords and punctuation
text = [word for word in text.split(' ') if word.isalpha()]

sw_english = nltk.corpus.stopwords.words('english')
text = [word for word in text if word not in sw_english]
```

```python
stemmer = SnowballStemmer('english')
stem_text = [stemmer.stem(word) for word in text]
stem_text
```

```python
lemmatizer = WordNetLemmatizer()
lemm_text = [lemmatizer.lemmatize(word) for word in text]
lemm_text
```

```python
# frequency distribution
list(set[lemm_text])[:10]
```

```python
# Find a frequency distribution of the words in the text
fdist = FreqDist(lemm_text)
fdist
```

```python
# provide the 10 most used words
used_words = fdist.most_common(10)
used_words
```

```python
# Find the 10 most used words that contain more than 3 letters
used_words_3 = [word for word in fdist.most_common(10) if len(word[0]) > 3]
used_words_3
```
